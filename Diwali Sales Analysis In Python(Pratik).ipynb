# EDA

```python
# EDA IN GENDER
```

```python
df.info()
```

```python
fx = sns.countplot(x = 'Gender' , data = df)

for i in fx.containers:
    fx.bar_label(i)
```

```python
# now we need to find which gender is spending more money
sales_gen = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(x = 'Gender',y= 'Amount' ,data = sales_gen)
```

# EDA IN  AGE

```python
ax=sns.countplot(data = df , x =  'Age Group' , hue = 'Gender')

for i in ax.containers:
    ax.bar_label(i)
```

```python
sales = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)

sns.barplot(x = 'Age Group' , y = 'Amount' , data = sales)
```

```python
df.info()
```

# EDA IN STATES

```python
state_order = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False)

sns.set(rc={'figure.figsize' : (25,15)})

sns.barplot(x= 'State' , y =   'Orders' ,data = state_order)
```

```python
state_amount = df.groupby(['State'] , as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)


sns.barplot(x='State' , y = 'Amount' , data = state_amount)
```

# EDA IN MARITAL STATUS

```python
sns.set(rc={'figure.figsize': (5,5)})
ax = sns.countplot(x='Marital_Status' , data= df)

for i in ax.containers:
    ax.bar_label(i)
```

```python
sale_marital_status = df.groupby(['Marital_Status' , 'Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.set(rc={'figure.figsize':(6,5)})

sns.barplot(x= 'Marital_Status' , y = 'Amount',hue = 'Gender' , data = sale_marital_status)
```

# EDA IN OCCUPATION

```python
ax = sns.countplot(x ='Occupation' , data = df )

sns.set(rc ={'figure.figsize': (20,20)})

for i in ax.containers:
    ax.bar_label(i)
```

```python
sale_amount = df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.set(rc= {'figure.figsize': (20,5)})

sns.barplot(x= 'Occupation' , y = 'Amount' , data = sale_amount )
```

# EDA IN PRODUCT

```python
ax = sns.countplot( x='Product_Category' , data= df)

sns.set(rc={'figure.figsize':  (28, 5)})
```

```python
sale_product_amount = df.groupby(['Product_Category'] , as_index=False)['Amount'].sum().sort_values(by='Amount' , ascending = False)

sns.set(rc={'figure.figsize': ( 30 , 10)})

sns.barplot( x= 'Product_Category' , y = 'Amount' , data= sale_product_amount)
```

```python
sale_product_amount = df.groupby(['Product_Category'] , as_index=False)['Orders'].sum().sort_values(by='Orders' , ascending = False)

sns.set(rc={'figure.figsize': ( 30 , 10)})

sns.barplot( x= 'Product_Category' , y = 'Orders' , data= sale_product_amount)
```
